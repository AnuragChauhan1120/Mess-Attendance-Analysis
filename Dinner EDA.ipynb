```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("dinner_clean_updated.csv.xlsx") 
```

```python
df = df.replace({np.nan: None})
df
```

```python
print("\nAttendance Stats:\n", df["Attendance"].describe())
```

```python
attendance_by_day = df.groupby("Day")["Attendance"].mean()
attendance_by_weather = df.groupby("Weather")["Attendance"].mean()
attendance_by_exam = df.groupby("Exam Week")["Attendance"].mean()
attendance_by_event = df.groupby("Special Event")["Attendance"].mean()
```

```python
print(attendance_by_day)
```

```python
print(attendance_by_weather)
```

```python
print(attendance_by_exam)
```

```python
print(attendance_by_event)
```

```python
plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
attendance_by_day.plot(kind='bar', color='skyblue')
plt.title("Average Attendance by Day")
plt.ylabel("Attendance")
```

```python
plt.subplot(2, 2, 3)
attendance_by_exam.plot(kind='bar', color='green')
plt.title("Average Attendance by Exam Week")
plt.ylabel("Attendance")
```

```python
plt.subplot(2, 2, 4)
attendance_by_event.plot(kind='bar', color='purple')
plt.title("Average Attendance by Event")
plt.ylabel("Attendance")
```
